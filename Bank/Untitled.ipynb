```python
# Check if 'color column exists in the arquivoFrame
if 'sex' in arquivo.columns:
    arquivo['car'] = arquivo['car'].replace('NO', 0)
    arquivo['car'] = arquivo['car'].replace('YES', 1)
else:
    print("DEU RUIM")

# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
# Check if 'color column exists in the arquivoFrame
if 'sex' in arquivo.columns:
    arquivo['save_act'] = arquivo['save_act'].replace('NO', 0)
    arquivo['save_act'] = arquivo['save_act'].replace('YES', 1)
else:
    print("DEU RUIM")

# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
# Check if 'color column exists in the arquivoFrame
if 'sex' in arquivo.columns:
    arquivo['current_act'] = arquivo['current_act'].replace('NO', 0)
    arquivo['current_act'] = arquivo['current_act'].replace('YES', 1)
else:
    print("DEU RUIM")

# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
# Check if 'color column exists in the arquivoFrame
if 'sex' in arquivo.columns:
    arquivo['mortgage'] = arquivo['mortgage'].replace('NO', 0)
    arquivo['mortgage'] = arquivo['mortgage'].replace('YES', 1)
else:
    print("DEU RUIM")

# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
# Check if 'color column exists in the arquivoFrame
if 'sex' in arquivo.columns:
    arquivo['pep'] = arquivo['pep'].replace('NO', 0)
    arquivo['pep'] = arquivo['pep'].replace('YES', 1)
else:
    print("DEU RUIM")

# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
# Check if 'color column exists in the arquivoFrame
if 'sex' in arquivo.columns:
    arquivo['region'] = arquivo['region'].replace('INNER_CITY', 0)
    arquivo['region'] = arquivo['region'].replace('TOWN', 1)
    arquivo['region'] = arquivo['region'].replace('RURAL', 2)
    arquivo['region'] = arquivo['region'].replace('SUBURBAN', 3)
else:
    print("DEU RUIM")

# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
arquivo = arquivo.drop('id', axis = 1)
# Display the first few rows of the arquivoFrame
print(arquivo.head())
```

```python
for marr in arquivo["married"]:
    if marr == 1:
        arquivo["percapita"] =  arquivo["income"] /(arquivo["children"] + 2)
    else:
        arquivo["percapita"] =  arquivo["income"] /(arquivo["children"] + 1)
    

print(arquivo.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
arquivo_scaled = scaler.fit_transform(arquivo)

kmeans = KMeans(n_clusters=3, random_state=42, n_init = 10)
kmeans.fit(arquivo_scaled)
clusters = kmeans.predict(arquivo_scaled)

plt.scatter(arquivo['percapita'], arquivo['pep'], c=clusters)
plt.xlabel('Idade')
plt.ylabel('Renda')
plt.title('Clusters de clientes')
plt.show()
```

```python
y = arquivo['pep']
X = arquivo.drop('pep', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

# Assuming x is your feature set and y is the target variable
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Importando StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Escalamento dos dados
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
# Para dados com features contnuas que seguem uma distribuio aproximadamente normal
modelo = make_pipeline(StandardScaler(), SVC(probability=True))

modelo.fit(x_train, y_train)
y_pred = modelo.predict(x_test)
```

```python
resultado = modelo.score(x_test, y_test)
print ("Acurcia:", resultado)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
# Adicionando a verificao da Preciso, Recall e F1-Score
from sklearn.metrics import precision_score, recall_score, f1_score

# Fazendo previses no conjunto de teste
y_pred = modelo.predict(x_test)

"""
Calculando e exibindo a matriz de confuso. A orientao padro  a seguinte:
[0,0]: Verdadeiros Negativos (VN) - Previses corretamente identificadas como negativas.
[0,1]: Falsos Positivos (FP) - Previses incorretamente identificadas como positivas.
[1,0]: Falsos Negativos (FN) - Previses incorretamente identificadas como negativas.
[1,1]: Verdadeiros Positivos (VP) - Previses corretamente identificadas como positivas.
"""

conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confuso:")
print(conf_matrix)
```

```python
"""
Calculando e exibindo as mtricas de classificao.
Se algumas classes tm muito mais amostras do que outras, isso pode influenciar o desempenho e confiabilidade do modelo.

O "support" refere-se  quantidade de ocorrncias da classe especfica no conjunto de dados, sendo til para verificar desbalanceamentos.
A "macro avg" calcula a mdia aritmtica das mtricas (preciso, recall, F1-score) para cada classe, sem considerar o nmero de instncias em cada classe (support).
A "weighted avg" calcula a mdia ponderada das mtricas para cada classe, considerando o nmero de instncias em cada classe (support).
"""
print("Relatrio de Classsificao:")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Prever as probabilidades para os dados de teste.
# Note que estamos interessados nas probabilidades da classe positiva (1), ento usamos [:, 1].
y_probs = modelo.predict_proba(x_test)[:, 1]

# Calcular FPR, TPR, e limiares
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calcular a AUC
roc_auc = auc(fpr, tpr)

# Plotar a curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
